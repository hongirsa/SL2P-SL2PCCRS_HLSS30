 ## renname 

```python
# rename filename from 'sitesS2SR_FO_2007_2014_mixed_2019_apr_nov_fraction50_cloud.csv' to 'S2_FO_2007_2014_mixed_2019_SR.csv'
for path, dirs, files in os.walk("E:/vegetation/gee/notebook/CFS/HW_SW/S2/SR"):
        # print (files)
        print (path)
        # print (dirs)
        for filename in files:            
            temp1=filename.replace('_apr_nov_fraction50_cloud.csv', '')
            temp2=temp1.replace('sitesS2SR_', '')
            temp3='S2_'+temp2+'_SR.csv'            
            print (temp3)
            rename(path+'/'+filename,path+'/'+temp3)
            # temp3= temp2
```

```python
# rename filename from 'sitesS2SR_FO_2007_2014_mixed_2019_apr_nov_fraction50_cloud_fCOVER_SL2P2.csv' to 'S2_FO_2007_2014_mixed_2019_fCOVER.csv' same for LAI
for path, dirs, files in os.walk("E:/vegetation/gee/notebook/CFS/HW_SW/S2/BioPhy_para"):
        # print (files)
        print (path)
        # print (dirs)
        for filename in files:            
            temp1=filename.replace('_apr_nov_fraction50_cloud', '')
            temp2=temp1.replace('sitesS2SR_', '')
            temp3=temp2.replace('_SL2P2','')
            temp4='S2_'+temp3            
            print (temp4)
            rename(path+'/'+filename,path+'/'+temp4)
            # temp3= temp2
```
